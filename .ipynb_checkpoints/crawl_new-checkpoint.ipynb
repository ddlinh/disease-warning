```python
import pandas as pd
import numpy as np
from selenium import webdriver
from bs4 import BeautifulSoup
import re
import time
```

```python
driver = webdriver.Firefox()
driver.get('https://thoitietvietnam.locvy.com/thoi-tiet-hang-ngay/binh-dinh')
soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
day_sources = soup.find_all('div', {'class':'panel-body daily'})
days = []
temps = []
rain_probas = []
rains = []
gusts = []
humids = []
uv_nums = []
uv_labels = []

def add_data(day_num, day_source):
    for idx, ele in enumerate(day_source):
        if (ele.name == 'div' and ele.string is not None): # or i.name == 'hr':
    #         print(ele.string, ele.name, idx)
            if idx == 5:
                days.append(day_num)
                temp = int(re.search('Cao \d+',ele.string).group().split()[-1])
                rain_proba = int(re.search('Khả năng có mưa \d+',ele.string).group().split()[-1])
                rain = re.search('Lượng mưa khoảng \d+',ele.string)
                if rain is None and rain_proba < 60:
                    rain = 0
                elif rain is not None:
                    rain = int(rain.group().split()[-1])
                else:
                    rain = None
                temps.append(temp)
                rain_probas.append(rain_proba)
                rains.append(rain)
#                 print('temp: {}, rain_proba: {}, rain: {}'.format(temp, rain_proba, rain))
            if idx == 9:
                gust = int(re.search('Tốc độ gió: \d+',ele.string).group().split()[-1])
                gusts.append(gust)
#                 print('gust: {}'.format(ele.string))
            if idx == 11:
                humid = int(re.search('Độ ẩm: \d+',ele.string).group().split()[-1])
                humids.append(humid)
#                 print('humid: {}'.format(ele.string))
            if idx == 13:
                uv_num = int(ele.string.split(': ')[-1])
                uv_nums.append(uv_num)
#                 print('uv_num: {}'.format(ele.string))
            if idx == 15:
                uv_label = ele.string.split(': ')[-1]
                uv_labels.append(uv_label)
#                 print('uv_label: {}'.format(ele.string))
            if idx == 23:
                days.append(day_num)
                temp = int(re.search('Thấp \d+',ele.string).group().split()[-1])
                rain_proba = int(re.search('Khả năng có mưa \d+',ele.string).group().split()[-1])
                rain = re.search('Lượng mưa khoảng \d+',ele.string)
                if rain is None and rain_proba < 60:
                    rain = 0
                elif rain is not None:
                    rain = int(rain.group().split()[-1])
                else:
                    rain = None
                temps.append(temp)
                rain_probas.append(rain_proba)
                rains.append(rain)
#                 print('temp: {}, rain_proba: {}, rain: {}'.format(temp, rain_proba, rain))
            if idx == 27:
                gust = int(re.search('Tốc độ gió: \d+',ele.string).group().split()[-1])
                gusts.append(gust)
#                 print('gust: {}'.format(ele.string))
            if idx == 29:
                humid = int(re.search('Độ ẩm: \d+',ele.string).group().split()[-1])
                humids.append(humid)
#                 print('humid: {}'.format(ele.string))
            if idx == 31:
                uv_num = int(ele.string.split(': ')[-1])
                uv_nums.append(uv_num)
#                 print('uv_num: {}'.format(ele.string))
            if idx == 33:
                uv_label = ele.string.split(': ')[-1]
                uv_labels.append(uv_label)
#                 print('uv_label: {}'.format(ele.string))
```

```python
start = time.time()
for day_num, day_source in enumerate(day_sources):
    add_data(day_num, day_source)
time.time() - start
```

```python
data = pd.DataFrame({'days':days,'temps':temps,'rains':rains,'rain_probas':rain_probas,'gusts':gusts,
              'humids':humids,'uv_nums':uv_nums,'uv_labels':uv_labels})
data['rains'].fillna(np.mean(data['rains']), inplace=True)
data_gb = data.groupby('days').agg(['mean', 'max', 'min']).reset_index()
data_gb.columns = [x[0]+'_'+x[1] for x in data_gb.columns]
data_gb['days_'] = 1
data_gbgb = data_gb.groupby('days_').agg(['mean', 'max', 'min']).reset_index()
data_gbgb.columns = [x[0]+'_'+x[1] for x in data_gbgb.columns]
data_gbgb
```
